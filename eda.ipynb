```python
from pathlib import Path
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Draw
from rdkit import RDLogger
RDLogger.DisableLog('rdApp.*')

root_dir = list(Path("data/samples/").glob("*"))
root_dir = [x for x in root_dir if x.is_dir()]
csvs = [x / f"{x.stem}_1.csv" for x in root_dir]
dfs = [pd.read_csv(x) for x in csvs]
```

```python
for csv, df in zip(csvs, dfs):
    print(csv)
    smiles = df.sample(100)["SMILES"]
    mols = [Chem.MolFromSmiles(x) for x in smiles]
    mols = [x for x in mols if x is not None][0:10]
    display(Draw.MolsToGridImage(mols, molsPerRow=5))
```
